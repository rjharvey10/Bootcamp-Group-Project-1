```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# CSV file
flight_data_csv = "Air_Traffic_Passenger_Statistics.csv"

# Read the mouse data and the study results
flights_df = pd.read_csv(flight_data_csv)
flights_df.head()
```

```python
flights_df.dropna()
flights_df.describe()
```

```python
flights_df = flights_df.drop(['Operating Airline', 'Operating Airline IATA Code', 'Published Airline','Published Airline IATA Code','Price Category Code','Terminal','Boarding Area'], 1)
flights_df.head()
```

```python
flights2020_df = flights_df.loc[flights_df["Activity Period"] >= 202001]
flights2020_df.head()
```

```python
flightdata_df = flights2020_df.loc[flights2020_df["Activity Type Code"] == "Enplaned"]
flightdata_df.head()
```

```python
flightdata_grouped = flightdata_df.groupby(['Activity Period'])
flightdata_grouped.head()
```

```python
flightdata_sum = flightdata_grouped.sum()
flightdata_sum.head(20)
```

```python
flights_byregion = flightdata_df.groupby(['Activity Period'] + ['GEO Region'])
flights_byregion.head()
```

```python
flightcountry_grouped = flights_byregion.sum()
flightcountry_grouped.head(24)
```

```python
flights_intl = flightdata_df.groupby(['Activity Period'] + ['GEO Summary'])
flights_intl.head()
```

```python
flightsintl_sum = flights_intl.sum()
flightsintl_sum.head()
```
